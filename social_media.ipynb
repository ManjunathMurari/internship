```python
import pandas as pd
from matplotlib import pyplot as plt

# Load the CSV file
df = pd.read_csv("social_media.csv")

# Display first 5 rows
print(df.head())
```

```python
###  Bar plot  ###
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("social_media.csv")


platform_comments = df.groupby("platform")["comments"].sum()
colors=["r","b","g","y"]

plt.figure()
plt.bar(platform_comments.index, 
        platform_comments.values,
        color=colors,edgecolor="r",
        linewidth=2,
        linestyle=":",
        alpha=0.6)

plt.xlabel("Platform")
plt.ylabel("Total Comments in 10^7")
plt.title("Total Comments by Platform")

plt.show()

```

```python
###  pie plot ###

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("social_media.csv")


platform_likes = df.groupby("platform")["likes"].sum()
platform_comments = df.groupby("platform")["comments"].sum()
platform_shares = df.groupby("platform")["shares"].sum()

values=[platform_likes.sum(),platform_comments.sum(),platform_shares.sum()]
labels=["likes","comments","shares"]

plt.figure()
plt.pie(values,labels=labels,textprops={'fontsize': 12},autopct="%1.1f%%")
plt.show()
```

```python
### scatter plot , views vs likes ###

import pandas as pd
from matplotlib import pyplot as plt

df=pd.read_csv("social_media.csv")

platform_likes = df.groupby("platform")["likes"].sum()
platform_views = df.groupby("platform")["views"].sum()

total_likes=platform_likes.sum()
total_views=platform_views.sum()

plt.scatter(platform_views,platform_likes)
plt.show()

```

```python
### Heat map ###
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv("social_media.csv")

# Select numeric columns
data = df[["likes", "comments", "shares", "views"]]

# Compute correlation matrix
corr = data.corr()

# Create heat map
plt.figure()
plt.imshow(corr)

# Add color bar
plt.colorbar()

# Add axis labels
plt.xticks(np.arange(len(corr.columns)), corr.columns)
plt.yticks(np.arange(len(corr.columns)), corr.columns)

# Add values inside cells
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, f"{corr.iloc[i, j]:.2f}",
                 ha="center", va="center")

plt.title("Engagement Correlation Heatmap")
plt.show()
```
